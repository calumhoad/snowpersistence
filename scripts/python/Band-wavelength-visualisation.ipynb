```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the data from the CSV file
df = pd.read_csv('../data/band-wavelengths.csv')

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.barh(range(len(df)), df['STOP'] - df['START'], left=df['START'], height=0.5, color='blue')

# Customize the plot
plt.xlabel('Wavelength (nm)')
plt.title('Sensor Bands Range Plot')
plt.yticks(range(len(df)), df['LABEL'])

# Show the plot
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Sort the DataFrame by the 'start_wavelength' column
df = df.sort_values(by='START')

# Create a list of colors based on the band labels
colors = ['red' if label.startswith('MAIA') else 'blue' for label in df['LABEL']]

# Create a bar plot with colors
plt.figure(figsize=(10, 6))
for i, (start, stop, label) in enumerate(zip(df['START'], df['STOP'], df['LABEL'])):
    plt.barh(i, stop - start, left=start, height=0.5, color=colors[i])
    plt.text(start - 10, i, f'{start}', va='center', ha='right', color='black', fontsize=10)
    plt.text(stop + 10, i, f'{stop}', va='center', ha='left', color='black', fontsize=10)

# Customize the plot
plt.xlabel('Wavelength (nm)')
plt.title('Sensor Bands Range Plot')
plt.yticks(range(len(df)), df['LABEL'])

# Show the plot
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
