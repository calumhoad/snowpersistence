```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the data from the CSV file
df = pd.read_csv('../../data/sensor-info/band-wavelengths.csv')
```

```python
df
```

```python
# Sort the DataFrame by the 'START' column
df = df.sort_values(by='START')

# Create a list of colors based on the band labels
colors = ['#00A4CCFF' if label.startswith('MAIA') else '#F95700FF' for label in df['LABEL']]

# Create a bar plot with colors
plt.figure(figsize=(12, 6))
for i, (start, stop, label, name) in enumerate(zip(df['START'], df['STOP'], df['LABEL'], df['NAME'])):
    if label.startswith('MAIA'):
        label_text = 'MAIA'
    else:
        label_text = 'M3M'
    
    plt.barh(i, stop - start, left=start, height=0.5, color=colors[i])
    plt.text(start - 5, i, f'{label_text}', va='center', ha='right', color='black', fontsize=10, fontweight='bold')
    plt.text(stop + 5, i, f'{name}', va='center', ha='left', color='black', fontsize=10) #, fontweight='bold')

# Customize the plot
plt.xlabel('Wavelength (nm)')
#plt.title('Sensor Bands Range Plot')
#plt.yticks(range(len(df)), df['LABEL'])
plt.xlim(350, 950)
plt.yticks([])

# Show the plot
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
